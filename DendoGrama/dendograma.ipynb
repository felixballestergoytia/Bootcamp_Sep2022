```python
##
import pandas as pd
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
variables,etiquetas=make_blobs(n_samples=10,centers=3,cluster_std=2.5)
```

```python
plt.scatter(variables[:,0],variables[:,1]);
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
Enlances=linkage(variables,'single')
```

```python
Enlances
```

```python
dendrogram(Enlances,orientation="top",labels=range(1,11))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster=AgglomerativeClustering(n_clusters=4)
```

```python
cluster.fit_predict(variables)
```

```python
valores_1,etiquetas_1=make_blobs(n_samples=500,centers=4,cluster_std=2)
```

```python
Enlances=linkage(valores_1,'single')
```

```python
dendrogram(Enlances,orientation="top",labels=range(1,501))
```

```python
## Metodo de silueta
clusters=range(2,10)
from sklearn.metrics import silhouette_score
```

```python
silueta=[]
for x in clusters:
    modelo = AgglomerativeClustering(
        n_clusters=x,affinity="cosine",linkage="complete"
    )
    etiquetas_cluster=modelo.fit_predict(valores_1)
    silueta.append(silhouette_score(valores_1,etiquetas_cluster))
```

```python
silueta
```

```python
plt.plot(clusters,silueta,marker="o")
plt.title("Indice silueta")
plt.xlabel("Nmero de cluster")
plt.ylabel("Indice silueta")
```
